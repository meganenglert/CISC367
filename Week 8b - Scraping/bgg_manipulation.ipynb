```python
import pandas as pd
df =pd.read_json("bgg\\bgg-all.json")
df
```

```python
df.dtypes
```

```python
df = df[df['rank']!= 'N/A']
df
```

```python
df['rank'] = pd.to_numeric(df['rank'])
df.year = pd.to_numeric(df['year'])
df.geek_rating = pd.to_numeric(df['geek_rating'])
df.avg_rating = pd.to_numeric(df.avg_rating)
df.num_voters = pd.to_numeric(df.num_voters)
df.dtypes
```

```python
df.set_index(df['rank'])
```

```python
df.to_csv("bgg_ranks.csv",index = False)
```
